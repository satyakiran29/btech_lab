6. Write a Python program to compute the conditional probability P(A|B), where:
A = "Student passed Math"
B = "Student passed both Math and English"
Use a dataset or assume a sample of 100 students with passed subjects information.

```python
import pandas as pd

# Sample dataset of 100 students with passed subjects information

data = {
    'Student': [f'Student_{i+1}' for i in range(100)],
    'Passed_Math': np.random.choice([True, False], size=100, p=[0.7, 0.3]),
    'Passed_English': np.random.choice([True, False], size=100, p=[0.6, 0.4])
}

df_students = pd.DataFrame(data)

# Add a column to indicate if a student passed both Math and English
df_students['Passed_Both'] = df_students['Passed_Math'] & df_students['Passed_English']

# Calculate the probabilities
total_passed_both = df_students['Passed_Both'].sum()
total_passed_math = df_students['Passed_Math'].sum()

# Conditional probability P(A|B) = P(A and B) / P(B)
conditional_probability = total_passed_both / total_passed_math if total_passed_math > 0 else 0

print(f"Conditional Probability P(A|B): {conditional_probability:.2f}")
```

17. Two independent sample have the following characteristics: Mean (Â¯x1) =40.5, standard
deviation(s1) =45, sample size(n1) = 50. Mean (Â¯x2) =38.2, standard deviation(s2) =5.0,
sample size(n2) =60. Test the null hypothesis that the two-population means are equal
(H0: Î¼1 = Î¼2) against the alternative hypothesis that they are not equal (H1: Î¼1 â‰  Î¼2)
using at 5% of level of significance

```python
from scipy.stats import t

# Given data
mean1 = 40.5
std_dev1 = 45
n1 = 50

mean2 = 38.2
std_dev2 = 5.0
n2 = 60

# Calculate the pooled standard error
pooled_se = ((std_dev1**2 / n1) + (std_dev2**2 / n2))**0.5

# Calculate the t-statistic
t_statistic = (mean1 - mean2) / pooled_se

# Degrees of freedom (approximation using Welch-Satterthwaite equation)
df = ((std_dev1**2 / n1) + (std_dev2**2 / n2))**2 / (
    ((std_dev1**2 / n1)**2 / (n1 - 1)) + ((std_dev2**2 / n2)**2 / (n2 - 1))
)

# Critical t-value for a two-tailed test at 5% significance level
alpha = 0.05
t_critical = t.ppf(1 - alpha / 2, df)

# p-value
p_value = (1 - t.cdf(abs(t_statistic), df)) * 2

# Print results
print(f"t-statistic: {t_statistic:.4f}")
print(f"Degrees of freedom: {df:.4f}")
print(f"Critical t-value: Â±{t_critical:.4f}")
print(f"p-value: {p_value:.4f}")

# Decision
if abs(t_statistic) > t_critical:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

18. Two samples have the following characteristics: variance(s12) =12.057, sample size(n1)
=8. variance(s22) =11.4, sample size(n2) =10. Test the null hypothesis that the two
population that the two population variances are equal (H0: Ïƒ12= Ïƒ22) against the
alternative hypothesis that they are not equal (H1: Ïƒ12â‰  Ïƒ22) using at 5% of level of
significance

```python
from scipy.stats import f

# Given data
variance1 = 12.057
n1 = 8
variance2 = 11.4
n2 = 10

# Calculate the F-statistic
f_statistic = variance1 / variance2 if variance1 > variance2 else variance2 / variance1

# Degrees of freedom
df1 = n1 - 1
df2 = n2 - 1

# Critical F-values for a two-tailed test at 5% significance level
alpha = 0.05
f_critical_low = f.ppf(alpha / 2, df1, df2)
f_critical_high = f.ppf(1 - alpha / 2, df1, df2)

# Decision
if f_statistic < f_critical_low or f_statistic > f_critical_high:
    decision = "Reject the null hypothesis (H0)."
else:
    decision = "Fail to reject the null hypothesis (H0)."

# Print results
print(f"F-statistic: {f_statistic:.4f}")
print(f"Degrees of freedom: df1={df1}, df2={df2}")
print(f"Critical F-values: [{f_critical_low:.4f}, {f_critical_high:.4f}]")
print(decision)
```

19. A sample of 10 observations has a mean of 0.742 the population mean is known to be
0.700 and the population standard deviation is 0.040 Test the null hypothesis that the
sample mean is equal to the population mean against the alternative hypothesis that the
sample mean is greater than the population mean, using a 5% level of significance

```python
from scipy.stats import norm

# Given data
sample_mean = 0.742
population_mean = 0.700
population_std_dev = 0.040
sample_size = 10
alpha = 0.05  # Level of significance

# Calculate the z-score
z_score = (sample_mean - population_mean) / (population_std_dev / (sample_size ** 0.5))

# Calculate the critical z-value for a one-tailed test
z_critical = norm.ppf(1 - alpha)

# Calculate the p-value
p_value = 1 - norm.cdf(z_score)

# Print the results
print(f"Z-Score: {z_score:.4f}")
print(f"Critical Z-Value: {z_critical:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision
if z_score > z_critical:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

20. A researcher collected a sample of 10 observations. He wants to test whether these
observations same from a normal distribution with a mean equal to the sample mean
perform chi square goodness to fit test to determine if the observations fit the specified
distribution.

```python
from scipy.stats import chisquare, norm
import numpy as np

# Sample data
sample_data = [12, 15, 14, 10, 13, 11, 14, 12, 13, 15]

# Calculate the sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Define the bins for the chi-square test
num_bins = 5
bins = np.linspace(min(sample_data), max(sample_data), num_bins + 1)

# Calculate observed frequencies
observed_frequencies, _ = np.histogram(sample_data, bins=bins)

# Calculate expected frequencies using the normal distribution
expected_frequencies = []
for i in range(len(bins) - 1):
    lower_bound = bins[i]
    upper_bound = bins[i + 1]
    prob = norm.cdf(upper_bound, loc=sample_mean, scale=sample_std) - norm.cdf(lower_bound, loc=sample_mean, scale=sample_std)
    expected_frequencies.append(prob * len(sample_data))

# Normalize expected frequencies to match the sum of observed frequencies
expected_frequencies = np.array(expected_frequencies)
expected_frequencies *= observed_frequencies.sum() / expected_frequencies.sum()

# Perform the chi-square goodness-of-fit test
chi_square_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Print the results
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The data does not fit the specified normal distribution.")
else:
    print("Fail to reject the null hypothesis: The data fits the specified normal distribution.")
```

24.Write a Python program to visualize the correlation between two variables using a scatter
plot. Use matplotlib to plot the data points and display the correlation coefficient on the
plot.

```python
# Example data for X and Y
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.2, 4.8, 7.5, 11.3, 15.1])

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(X, Y)[0, 1]

# Create a scatter plot
plt.scatter(X, Y, color='blue', label='Data Points')

# Add a trendline
coefficients = np.polyfit(X, Y, 1)
trendline = np.poly1d(coefficients)
plt.plot(X, trendline(X), color='red', label='Trendline')

# Add labels, title, and legend
plt.title("Scatter Plot with Correlation Coefficient")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Display the correlation coefficient on the plot
plt.text(min(X), max(Y), f"Correlation Coefficient: {correlation_coefficient:.2f}", fontsize=10, color='green')

# Show the plot
plt.grid()
plt.show()
```
